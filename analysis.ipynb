```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sleep_timings = [0, 1, 5, 10, 50, 100, 500, 1000, 5000, 10000]
result_percentile = []

for time in sleep_timings:
    filename = 'results/1M-' + str(time) + 'ns.txt'
    array = np.loadtxt(filename)
    result_percentile.append(np.percentile(array, 99))

plt.plot(sleep_timings, result_percentile)
plt.xscale('log')
plt.xlabel('time to sleep after every write, nsec')
plt.ylim(0, 1000)
plt.ylabel('99 percentile of write-read delay, nsec')
plt.show()
```

Как видно из результатов, они не совсем зависят от частоты записи и 99 перцентиль изменяется в пределах погрешности в 200нс. Но при этом лучшее измерение было при задержке между записями в 50нс, что соответствует 20млн записей в секунду.
